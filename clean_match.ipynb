```python
import pandas as pd
df_matches = pd.read_csv("data/matches.csv")
```

```python
df_matches.info()
```

```python
df_matches["round"].value_counts()
```

```python
# Suppression des matchs qui ne nous interessent pas 
clean_round = df_matches[df_matches["round"].str.contains("PRELIMINARY", na = False)].index
df_matches = df_matches.drop(clean_round)
```

```python
df_matches["round"].value_counts()
```

```python
# Renomme les matchs avec les bonnes nominations 
df_matches["round"] = df_matches["round"].replace(to_replace={"GROUP_STAGE":"Poules", "1/8_FINAL": "Huitièmes de finale","FIRST": "Huitièmes de finale", "1/4_FINAL" : "Quarts de finale", "QUARTERFINAL_STAGE": "Quarts de finale", "1/2_FINAL" : "Demi-finales", "SEMIFINAL_STAGE": "Demi-finales", "_FINAL" : "Finale", "FINAL_ROUND":"Poule Finale", "PLACES_3&4" : "Troisième place"  })
```

```python
# match apres 2010 
df_matches[df_matches["year"] > 2010].head(10)["year"].value_counts()
```

```python
# Pas de scores pour 2014
df_matches[df_matches["year"] > 2010].head(10)
```

```python
# donc on drop ces lignes :
df_matches = df_matches.drop(df_matches[df_matches["year"] > 2010].index)
```

```python
df_matches.head(1)
```

```python
#Colonne Score : 1er valeur -> score de fin de match, 2e valeur -> score à la mi temps. on veut garder seulement le score final 
df_matches["score final"] = df_matches["score"].str.split(" ").str[0] 
df_matches.head(1)
```

```python
# Attribuer le nombre de buts a chaque équipes : 
df_matches["home_result"] = df_matches["score final"].str.split("-").str[0]
df_matches["away_result"] = df_matches["score final"].str.split("-").str[1]
df_matches.head(1)
```

```python
#Suppression colonnes dont on a plus besoin 
df_matches = df_matches.drop(columns=["url", "score", "score final"])
```

```python
# Renomme les colonnes
df_matches = df_matches.rename(columns = {"team1" : "home_team", "team2": "away_team", "year":"date"})
```

```python
# Creation colonne resultats : 
def function(row): 
    if row["home_result"]> row["away_result"]:
        return "home_team"
    elif  row["home_result"]< row["away_result"] :
        return "away_home"
    else : 
        return "draw"

df_matches["result"] = df_matches.apply(function, axis=1)
```

```python
df_matches.head(3)
```

```python
# On recupere la ville depuis la colonne venue et on enlève le point en fin de mot
df_matches["city"] = df_matches["venue"].str.replace('.','')
```

```python
# On recupere la date depuis la colonne edition en ne gardant que ce qu'il y a avant le "-"
df_matches["edition"] = df_matches["edition"].str.split("-").str[0]
df_matches = df_matches.drop(columns='venue')
df_matches.head(3)
```

```python
# On met une majuscule a chaque debut de mots
df_matches["city"] = df_matches["city"].str.title()
```

```python
df_matches["home_team"].value_counts()
```

```python
# regex pour enlever les noms d'origine des pays entre () : 
#test 
import re
test = "New Zealand (Aotearoa)"
re.findall(r"^[^(]+", test)
```

```python
df_matches["home_team"] = df_matches["home_team"].str.extract(r"^([^(]+)")
df_matches["home_team"].value_counts()
```

```python
df_matches["away_team"] = df_matches["away_team"].str.extract(r"^([^(]+)")
```

```python
df_matches.sample(10)
```

```python
df_matches["city"].value_counts()
```

```python
# On retire les potentiels espaces 
df_matches["home_team"].str.strip()
df_matches["away_team"].str.strip()
df_matches["city"].str.strip()
```

```python
df_matches['date'] = df_matches['date'].astype(str) + '-01-01'
df_matches['date'] = pd.to_datetime(df_matches['date']).dt.date
```
