```python
import json
import pandas as pd
with open("data/data_2018.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
data
```

```python
# Le json a 5 clés 
data.keys()
```

```python
#1er dictionnaire : stadium (utile pour recuperer le nom des villes)
df_stadiums = pd.DataFrame(data["stadiums"])
df_stadiums
```

```python
df_stadiums = df_stadiums[["id", "city"]]
df_stadiums
```

```python
#2eme dictionnaire : Tvchannels  
df_tvchannels = pd.DataFrame(data["tvchannels"])
df_tvchannels.head(1)
```

```python
#3e dictionnaire : Teams (utile pour récuperer le nom des villes participantes) 
df_teams = pd.DataFrame(data["teams"])
df_teams.head(3)
```

```python
df_teams = df_teams[["id", "name"]]
df_teams
```

```python
#4e dictionnaire : Groups  (poules) -> dictionnaires imbriqués  
data["groups"]
data["groups"].keys()
```

```python
groups = ["a", "b", "c", "d", "e", "f", "g", "h"]

liste = []
for i in groups : 
    df = pd.DataFrame(data["groups"][i]["matches"])
    df["groups"] = i
    liste.append(df)


df_groups = pd.concat(liste)
df_groups
```

```python
#5e dictionnaire : Knockout (type de matches) -> dictionnaires imbriqués  
data["knockout"]
```

```python
data["knockout"].keys()
#Il y a 5 dictionnaires à l'interieur
```

```python
data["knockout"]["round_16"]
```

```python
#On converti chaque dictionnaires imbriqués en Df et on ajoute une colonne de reference(round) pour savoir de quel Df il vient)
round_16 = data["knockout"]["round_16"]["matches"]
df_round_16 = pd.DataFrame(round_16)
df_round_16["round"] = "Huitième de finale"
df_round_16.head(5)
```

```python
#De meme pour les 4 autres 
df_round_8 = pd.DataFrame(data["knockout"]["round_8"]["matches"])
df_round_8["round"] = "Quarts de finale"
df_round_4 = pd.DataFrame(data["knockout"]["round_4"]["matches"])
df_round_4["round"] = "Demi-finales"
df_round_2_loser = pd.DataFrame(data["knockout"]["round_2_loser"]["matches"])
df_round_2_loser["round"] = "Troisième place"
df_round_2 = pd.DataFrame(data["knockout"]["round_2"]["matches"])
df_round_2["round"] = "Finale"
```

```python
#On concatène les 5 df ensemble pour avoir l'ensemble des compétitions 
df_round = pd.concat([df_round_16,df_round_8, df_round_4,df_round_2_loser,df_round_2])
```

```python
#On reset les index pour ne pas avoir de doublons
df_round.reset_index(drop=True,inplace=True)
df_round.head(5)
```

```python
#Suppression des colonnes qui ne nous sont pas utiles
df_round = df_round.drop(columns=["name", "home_penalty", "away_penalty", "winner", "channels", "finished", "matchday", "type"])
```

```python
df_round.head(3)
```

```python
df_groups.head(3)
```

```python
#On réarrange le df groups (poules) pour pouvoir le concatecter avec le df round 
df_groups = df_groups.drop(columns=["name", "channels", "finished", "matchday", "groups"])
df_groups = df_groups.rename(columns={"type":"round" })
df_groups["round"] = df_groups["round"].replace(to_replace={"group" : "Poules"})
df_groups.head(1)
```

```python
#on concatène le df round et le df groups
df = pd.concat([df_round,df_groups])
```

```python
#converti la date pour n'avoir que l'année, le mois et le jour
df["date"] = pd.to_datetime(df["date"], utc=True).dt.date
```

```python
# Donc on merge avec le df Teams pour recuperer les noms des team 
df = df.merge(df_teams, how="left", left_on="home_team", right_on ="id" )
df = df.merge(df_teams, how="left", left_on="away_team", right_on ="id" )
df.head(2)
```

```python
df = df.drop(columns=["home_team", "away_team", "id_x", "id_y"])
```

```python
df = df.rename(columns={"name_x": "home_team", "name_y" : "away_team", })
```

```python
# City => utilise le df_stadium pour prendre l'id lié au stade
df = df.merge(df_stadiums, how="left", left_on="stadium", right_on ="id" )
df.head(2)
```

```python
df = df.drop(columns=["stadium","id"])
```

```python
# colonne resultats : 
def function(row): 
    if row["home_result"]> row["away_result"]:
        return "home_team"
    elif  row["home_result"]< row["away_result"] :
        return "away_home"
    else : 
        return "draw"

df["result"] = df.apply(function, axis=1)
```

```python
# edition : 
df["edition"] = pd.to_datetime(df["date"]).dt.year
```
